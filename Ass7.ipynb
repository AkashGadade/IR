# download image

```python
image = requests.get('https://www.shutterstock.com/image-photo/seen-behind-modern-business-woman-100nw-2266497009.jpg')

with open('image.jpg','wb') as file:
    for chunk in image.iter_content(chunk_size = 1024):
        file.write(chunk)
```

```python
import requests

# URL of the video you want to download
video_url = "http://commondatastorage.googleapis.com/gtv-videos-bucket/sample/ForBiggerFun.mp4"

# Send an HTTP GET request to the video URL
response = requests.get(video_url)

# Check if the request was successful
if response.status_code == 200:
    # Specify the local file path to save the video
    local_file_path = "downloaded_video.mp4"

    # Open the local file for writing in binary mode
    with open(local_file_path, 'wb') as video_file:
        # Write the content of the response to the local file
        video_file.write(response.content)

    print(f"Video downloaded and saved as {local_file_path}.")
else:
    print(f"Failed to retrieve the video. Status code: {response.status_code}")
```
